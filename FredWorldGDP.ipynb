### Step 3. Merge the GDP Data for Countries 
  - Convert country-specific series into df 
  - Join country-specific dfs into one
  - Extract the country names from df `title`

```
all_countries = []

try:
    for id in gdp_df.index:
        country_series = fred.get_series(series_id=id)
        country_df = country_series.to_frame(name=id)
        all_countries.append(country_df)
except Exception as e:
    print(e)
else:
    print('Data is ready!')
```

```
all_countries
```

```
# check the size of the list
len(all_countries)
```

```
# concatenate the data based on column value
all_countries_df = pd.concat(all_countries, axis=1)
all_countries_df
```

```
country1 = "UnitedStates" # country with higher GDP
country2 = "China" # country with lower GDP


query1 = f"""
CREATE OR REPLACE TEMPORARY VIEW v_{country1}_{country2}_stats
    AS SELECT date, {country1}, {country2}, ROUND({country2}/{country1}, 3) AS percentage
         FROM t_world_gdp
         WHERE date BETWEEN '2011-01-01' AND '2021-01-01'
         UNION ALL
         SELECT 'group_total', ROUND(SUM({country1}), 2), ROUND(SUM({country2}), 2), ROUND(SUM({country2})/SUM({country1}), 2)
         FROM t_world_gdp
         WHERE date BETWEEN '2011-01-01' AND '2021-01-01'
         UNION ALL
         SELECT 'overall_average', ROUND(AVG({country1}), 2), ROUND(AVG({country2}), 2), ROUND(AVG({country2})/AVG({country1}), 2)
         FROM t_world_gdp
         UNION ALL
         SELECT 'group_average', ROUND(AVG({country1}),2), ROUND(AVG({country2}),2), ROUND(AVG({country2})/AVG({country1}), 2)
         FROM t_world_gdp
         WHERE date BETWEEN '2011-01-01' AND '2021-01-01'
"""

query2 = f""" SELECT date,{country1}, {country2}, percentage FROM v_{country1}_{country2}_stats"""

query3 = f""" DROP VIEW IF EXISTS v_{country1}_{country2}_stats """

queries = [query1, query2, query3]

for q in queries:
   spark.sql(q).show()
```
